Introduction 

Our goal with this project is to answer the predictive question: "Can the category of the knowledge level of users (UNS) be predicted by two variables, degree of study time for goal object materials (STG) and their standardized exam performances for goal object (PEG), and if so, what is their relationship (whether it is linear or non-linear, positive or negative, and strong or weak)?". The data set for this model comes from "uns_data_xlsx". 

To predict UNS, we would develop a model with data on STG and PEG. We will run a classification analysis for the three variables (LPR, PEG and UNS) using the K-nearest neighbors algorithm. This is significant, because it would be able to show the relationship between the degree of study time and users' knowledge levels. If a strong relationship exists, we would learn 'what is best degree of hours students can study to acquire a higher understanding'. This is not only helpful for students striving to perform well in this field, but also inspire institutions or program developers to improve their assisatnce to optimize students' knowledge level. 

About the data set being used:

This data is about the students' knowledge status about the subject of Electrical DC Machines. Electrical DC Machines. Also known as a DC machine, the machine is a generator which converts mechanical energy into electrical energy. The data determines students’ knowledge level (UNS) with 5 variables: STG, SCG, STR, LPR and PEG. 

In our project, we will focus on the variables below. 

Response variable: 
* UNS: Categorical and scaled variable, represents the knowledge level of the users

Predictive variables:
* STG: The degree of study time for goal object materials
* PEG: Continuous and scaled variable, eam performance of the user for goal objects

Methods 

We will select variables STG, PEG and UNS. The predictors STG and PEG are continuous, and the response varible (UNS) is categorical. We split the data into training and testing set with a 75% to 25% on random and utilize k-nearest neighbors algorithm, 5-fold cross-validation on training set to do classification analysis. 

We will pick the optimal k-value with the highest accuracy and carry out predictions on UNS on the testing set. 

Then, we will re-evaluate the accuracy of the prediction model by computing its prediction accuracy. 

For visualization, ew will graph a scatter plot for UNS vs PEG and UNS vs STG. 

We will also show the confusion matrix to further evaluate its usefulness. 

```r
#loading packages
library(tidyverse)
library(repr)
library(readxl)
library(cowplot)
library(tidymodels)
library(RColorBrewer)
options(repr.matrix.max.rows = 6)
```

```r
#reading the data
uns_data <- read_excel("uns_test_data.xlsx")
uns_data
```

```r
#cleaning data by changing UNS into a categorical variable that was previously a character
uns_data_mutated <- mutate(uns_data, UNS = as_factor(UNS))
uns_data_mutated
```

```r
#cleaning data by selecting three variables of interest, LPR, PEG and UNS
uns_data_selected <- select(uns_data_mutated, STG, PEG, UNS)
uns_data_selected
```

```r
#splitting the data into training and testing data;
#75% of the data as training and 25% as testing set
#Using the strata argument for UNS to get approximately the same percentage of each class in both training and testing set

set.seed(100)

uns_split <- initial_split(uns_data_selected, prop = 0.75, strata = UNS)
uns_training <- training(uns_split)
uns_testing <- testing(uns_split) 

uns_split
uns_training
uns_testing
```

```r
#summarizing training data - total number of observations)
nrow(uns_training)
```

```r
#summarizing training data - mean of numeric variables, PEG and LPR
uns_data_summarize <- summarize(uns_training, 
                                mean_STG = mean(STG), 
                                mean_PEG = mean(PEG),)
uns_data_summarize
```

```r
#summarizing training data - categories/classes of UNS
uns_training |>
pull(UNS) |>
levels()
```

```r
#summarizing UNS - number of observations for each class & the percentage of them in terms of the total dataset
number_of_observations_in_total <- nrow(uns_data_selected)
uns_data_summarize <- uns_data_selected |>
group_by(UNS) |>
summarize(count = n(), 
          percentage = n() / number_of_observations_in_total * 100)
uns_data_summarize
```

```r
#summarizing training data - classes in percentage of overall data

number_of_observations <- nrow(uns_training)
uns_data_summarize <- uns_training |>
group_by(UNS) |> 
summarize(count = n(), 
          percentage = n() / number_of_observations * 100)
uns_data_summarize
```

From the code above, we found that the strata argument that created the initial split worked well because the percentages of each class to the training set were roughly equivalent to the overall data. 

```r
#visualizing data of UNS vs STR and UNS vs LPR (not standardized variables) 
#Purpose: examine whether there are relationships between the two predictive variables & two responsive variables

options(repr.plot.width = 12, repr.plot.height = 12)

uns_data_plot <- ggplot(uns_training, aes(x = STG, y = PEG, color = UNS, shape = UNS)) +
                        geom_point(alpha = 2) +
                        labs(x = "The degree of study time for goal object materials (STG)", 
                             y = "Exam performance of user for goal object (PEG)", 
                             color = "The knowledge level of user (UNS)") +
                        ggtitle("The knowledge level of user (UNS) VS The degree of study time for goal object materials (SCG) \n  and \n Exam performance") +
                        theme(text = element_text(size = 14))
uns_data_plot
```

From the plot above, we found that there is a  positive linear relationship between PEG and UNS. Nonetheless, it is difficult to identify a relationship between SCG and UNS, which appears to be have a weaker positive relationship.
